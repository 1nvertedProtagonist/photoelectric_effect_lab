```python
##Time to start plotting graphs. First, find the mean and SEM for each point of each run
##Then, perform a weighted mean of each run by each SEM and get the final plotabble points and their error bars, this is plottable

rawxdata = np.asarray(np.transpose([[1/577],[1/546],[1/436],[1/405],[1/365]]))
rawxerr = np.asarray(np.full((1,5),None))
rawydata = np.asarray(run1_vols[0:105,:]) 
rawyerr = np.asarray(np.full((105,5),0.0001))
#print(rawxdata.shape)
#print(rawxerr.shape) 
#print(rawydata.shape) 
#print(rawyerr.shape)


plottable_xdata = glt.find_plottabl_stuff(rawxdata, rawxerr)[0]# throw these into goated function
#print(plottable_xdata.shape)
plottable_xuncs = glt.find_plottabl_stuff(rawxdata,rawxerr)[1] #subject to comparison when plotting
#print(plottable_xuncs.shape)
plottable_ydata = glt.find_plottabl_stuff(rawydata,rawyerr)[0]#throw these into goated function
#print(plottable_ydata.shape)
plottable_yuncs = glt.find_plottabl_stuff(rawydata,rawyerr)[1]#subject to comparison when plotting
#print(plottable_yuncs.shape)



final_xunc = glt.uncer_compare(plottable_xuncs,rawxerr)
final_yunc = glt.uncer_compare(plottable_yuncs,rawyerr)[0:1,:]
#print(final_xunc.shape)
#print(final_yunc.shape)

def linear(x,m,b): 
  return m*x+b 

print(final_yunc.size)
print(final_yunc.shape)
print(plottable_ydata.size)
print(plottable_ydata.shape)

y=np.squeeze(plottable_ydata)
sigmA = np.squeeze(final_yunc)
params, cov_matrix = curve_fit(linear,plottable_xdata,y, sigma = sigmA, absolute_sigma=True)
m,b = params
errors = np.diag(cov_matrix)
m_err,b_err = errors
y_hat = linear(plottable_xdata,m,b)

x = np.asarray(plottable_xdata).ravel()
y = np.asarray(plottable_ydata).ravel()
sigma = np.asarray(final_yunc).ravel()

print("shapes:", x.shape, y.shape, sigma.shape)
print("finite:", np.isfinite(x).all(), np.isfinite(y).all(), np.isfinite(sigma).all())
print("sigma>0:", (sigma > 0).all())
print("any None:", np.any(x == None), np.any(y == None), np.any(sigma == None))

print("about to fit...")
try:
    params, cov = curve_fit(linear, x, y, sigma=sigma, absolute_sigma=True)
    print("fit done.")
    print("params:", params)
    print("cov diag:", np.diag(cov))
except Exception as e:
    print("curve_fit failed:", repr(e))
    


```
